```python
import numpy as np
import matplotlib.pylab as plt
from tqdm import tqdm
from scipy.stats import norm
from types import SimpleNamespace
import binascii

params = SimpleNamespace(t=128, m=13, n=129)

TRACEDIR="traces"
SAMPLES=287002932

# The number of PoIs we are looking for:

PK_NROWS = params.t * params.m

POIS = sum(range(PK_NROWS - 1, PK_NROWS - 1 - params.n, -1)) + (PK_NROWS - 1) * params.n

class McElieceKey(object):
    def __init__(self, fname=None):
        with open(fname, 'r') as fh:
            self.output = fh.read()
        self.loads(self.output)

    def loads(self, kvpairs):
        result = {}
        lines = kvpairs.split('\n')
        for line in lines:
            line = line.strip()
            if not line or line.startswith('#'):
                continue
            key, value = map(str.strip, line.split('=', 1))
            value = value.strip("'\"")
            if key not in ["kem"]:
                value = bytes.fromhex(value)
            setattr(self, key, value)
            
        # Some post-processing.
        nr_leakbits = (PK_NROWS - 1) * PK_NROWS
        self.leak = self.leak[-nr_leakbits:]

    def save(self, fname):
        with open(fname, "w") as fh:
            keys = ['kem', 'seed', 'pk', 'sk', 'ct', 'ss', 'leak']
            for key in keys:
                value = getattr(self, key)
                if key not in ["kem"]:
                    value = binascii.hexlify(value).decode().upper()
                fh.write("{} = '{}'\n".format(key, value))

class Traces(object):
    def __init__(self, name):
        self.name = name
        self.fname_base = os.path.join(TRACEDIR, name)
        self.fname = self.fname_base + ".npy"
        self.fname_nfo = self.fname_base + ".nfo"
        self.fname_avg = self.fname_base + "-ravg.npy"
        # welford's online m2: https://en.wikipedia.org/wiki/Algorithms_for_calculating_variance
        self.fname_wm2 = self.fname_base + "-rwm2.npy"
        self.count = int(np.load(self.fname))
        if os.path.exists(self.fname_nfo):
            self.key = McElieceKey(fname=self.fname_nfo)
        else:
            self.key = None
        self.avg = np.load(self.fname_avg)
        self.wm2 = np.load(self.fname_wm2)

    def trace_fname_base(self, idx):
        return "{}-t{:04}".format(self.fname_base, idx)

```

```python
# generated by 04-find-leak.sage
MISSING=-1
PERIOD_SHIFT=33
PERIOD_ZERO=732
def model_leakpos(r, c):
    if r == c:
        return MISSING
    pos = -315.0*c^2 + 2264764.75*c + 1047823.0
    pos += -5.75*(c%8)
    pos += PERIOD_ZERO*r
    if r > c:
        pos += PERIOD_SHIFT - PERIOD_ZERO
    return int(pos)
```

```python
# The approximate period of the signal we are looking for.
period = ceil(SAMPLES / POIS)
# We want to be within 25% of our approximate signal.
period_min, period_max = 0.75 * period, 1.25*period
# Just as a sanity check, we check that we have sufficient samples.
min_fft_samples = ceil(1/(1/period_max - 1/(period_max+1)))

# The first round of gaussian elimination is approximately this long.
fftcount = period * (PK_NROWS - 1) * 8
assert(fftcount > min_fft_samples)
```

```python
# Select the samples to transform.
sig = abs(stat)[:fftcount]
fft = abs(np.fft.rfft(sig))
fftfreq = np.fft.rfftfreq(int(fftcount))

# This acts as a simple band filter, that removes a couple of problems:
# 1. There are some high-amplitude signals at very low frequencies (including the DC offset)
# 2. There are some high-amplitude signals at very high frequencies (for example with period 7)
fftidx_lo = np.searchsorted(fftfreq, 1/period_max)
fftidx_hi = np.searchsorted(fftfreq, 1/period_min)
print("{} (T={}) {} (T={})".format(fftidx_lo, period_max, fftidx_hi, period_min))

# Prescience mode on. We peaked at the data manually.
period_diag = 630
period_zero = 732
fftidx_diag = np.searchsorted(fftfreq, 1/period_diag)
fftidx_zero = np.searchsorted(fftfreq, 1/period_zero)
print("Cheatcode: {} (T={}) {} (T={})".format(fftidx_diag, period_diag, fftidx_zero, period_zero))
plt.plot(1/(fftfreq[range(fftidx_lo, fftidx_hi)]), fft[fftidx_lo:fftidx_hi])
```

```python
peakpos = np.argsort(abs(fft))
peakpos = peakpos[(peakpos >= fftidx_lo) & (peakpos <= fftidx_hi)]
peakpos = np.flip(peakpos)
# Values below threshold are excluded, as are ZONE samples before and after a peak.
PEAK_THRESHOLD = fft.mean() + 9*fft.std()
PEAK_ZONE = 10
peaks = []
print("Searching for peaks above {} with max width {}".format(PEAK_THRESHOLD, 2*PEAK_ZONE+1))
def peak_period(peak):
    return floor(0.5+1/fftfreq[peak])
while peakpos.size > 0:
    idx = peakpos[0]
    if fft[idx] < PEAK_THRESHOLD:
        break
    peaks.append(idx)

    remove_lo = np.searchsorted(fftfreq, 1/(peak_period(idx)+PEAK_ZONE), side='left')
    remove_hi = np.searchsorted(fftfreq, 1/(peak_period(idx)-PEAK_ZONE))
    #print("Removing idx", peak_period(idx), peak_period(remove_hi), peak_period(remove_lo))
    #print("Removing idx", idx, remove_lo, remove_hi)
    peakpos = peakpos[(peakpos < remove_lo) | (peakpos > remove_hi)]
for peak in peaks:
    print("Found signal with period {} and strength {}".format(peak_period(peak), fft[peak]))
```

```python
assert(len(peaks) == 2)
PERIOD_ZERO, PERIOD_DIAG = [peak_period(idx) for idx in sorted(peaks)]
assert(PERIOD_DIAG == period_diag)
assert(PERIOD_ZERO == period_zero)
```

```python
# Values below threshold are excluded, as are ZONE samples before and after a peak.
POI_THRESHOLD = 13
POI_ZONE = 5

poipos = np.where(abs(stat) > POI_THRESHOLD)[0]
# Gymnastics to reorder poipos by the value in stat
stat_values_at_poipos = abs(stat[poipos])
sorted_indices = np.argsort(stat_values_at_poipos)
poipos = poipos[sorted_indices]
poipos = np.flip(poipos)

pois = []
print("Searching for pois above {} with max width {}".format(POI_THRESHOLD, 2*POI_ZONE+1))
pois_skip = np.zeros(SAMPLES, dtype='bool')
pbar = tqdm(range(poipos.size))
for i in pbar:
    idx = poipos[i]
    if pois_skip[idx]:
        continue
    if abs(stat[idx]) < POI_THRESHOLD:
        break
    if len(pois) % 1000 == 0:
        pbar.set_description("strength {:.2f}".format(round(abs(stat[idx]), 2)))
    pois.append(idx)
    r1 = max(0, idx - POI_ZONE)
    r2 = min(SAMPLES, idx + POI_ZONE)
    pois_skip[r1:r2+1] = True
    #poipos = poipos[(poipos < (idx - POI_ZONE)) | (poipos > (idx + POI_ZONE))]
pois = np.array(pois)
```

```python
MISSING=-1
def longest_evenly_spaced_sequence(indices, PERIOD, ALLOWED_MISSING=1, MIN_LENGTH=7):
    # Ensure the indices are sorted
    indices = np.sort(indices)
    
    all_seqs = []
    longest_seq = []
    
    i = 0
    k = 0
    total_indices = len(indices)
    with tqdm(total=len(indices)) as pbar:
        while i < len(indices):
            current_seq = [indices[i]]
            next_value = indices[i] + PERIOD
        
            allowed_missing = ALLOWED_MISSING
            last_idx = 0
            for j in range(i + 1, len(indices)):
                if indices[j] > next_value + 1:
                    allowed_missing -= 1
                    if allowed_missing < 0:
                        break
                    current_seq.append(MISSING)
                    next_value += PERIOD
                if indices[j] == next_value - 1 or indices[j] == next_value or indices[j] == next_value+1:
                    current_seq.append(indices[j])
                    last_idx = j
                    allowed_missing = ALLOWED_MISSING
                    next_value += PERIOD
                
            while current_seq[-1] == MISSING:
                current_seq = current_seq[:-1]

            if len(current_seq) > len(longest_seq):
                #pbar.set_description("length {}".format(len(current_seq)))
                longest_seq = current_seq
                if len(longest_seq) >= MIN_LENGTH: #5: #PK_NROWS-params.n:
                    #pbar.write("Found something exciting: {} length {}".format(longest_seq[0], len(longest_seq)))
                    #pbar.write("X {} {} {} {}".format(i, len(longest_seq), longest_seq[0], longest_seq[-1]))
                    i = last_idx
                    all_seqs.append(np.array(longest_seq))
                    longest_seq = []
            i = i + 1
            if k < 10:
                k = k + 1
            pbar.update(int(i - pbar.n))
    
    return all_seqs

MIN_LENGTH=7
seqs = longest_evenly_spaced_sequence(pois, PERIOD_ZERO, MIN_LENGTH=MIN_LENGTH)
```

```python
# Now verify that we collected all the bits.
idx = 0
seqs_before_if = []
seqs_after_if = []
leakpos = []
for col in range(params.n):
    seq = seqs[idx]
    if col < MIN_LENGTH:
        # For the first MIN_LENGTH columns, we do not expect to find the iterations before
        # the if-condition.
        seqs_before_if.append(np.array([MISSING]*col))
    else:
        # For later columns, we actually identify all iterations of the loop.
        assert(len(seq) == col)
        bits_missing = sum(seq == MISSING)
        if bits_missing > 0:
            print("Bits missing in column {} before if: {} {}".format(col, bits_missing, np.where(seq == MISSING)[0]))
        seqs_before_if.append(seq)
        idx = idx + 1
        seq = seqs[idx]
    assert(len(seq) == PK_NROWS - 1 - col)
    bits_missing = sum(seq == MISSING)
    if bits_missing > 0:
        print("Bits missing in column {} after if: {} {}".format(col, bits_missing, np.where(seq == MISSING)[0]))
    seqs_after_if.append(seq)
    idx = idx + 1
    leakpos.extend(seqs_before_if[-1])
    leakpos.extend(seqs_after_if[-1])
leakpos = np.array(leakpos)
expected_leakpos_len = (PK_NROWS - 1) * params.n
if (len(leakpos) != expected_leakpos_len):
    print("warning: did not get expected number of leaked bits:", len(leakpos), expected_leakpos_len)
assert(len(leakpos) == expected_leakpos_len)
# We expect at least MIN_LENGTH*(MIN_LENGTH-1)/2 missing elements.
nr_missing_expected = MIN_LENGTH*(MIN_LENGTH-1)/2
nr_missing = sum(leakpos == MISSING)
if nr_missing != nr_missing_expected:
    print("warning: some bits were missing:", nr_missing - nr_missing_expected)
```

```python
# We expect a small shift when the if condition is true.
# Measure this shift here.
before_and_after_deltas = []
for col in range(params.n):
    seq1 = seqs_before_if[col]
    seq2 = seqs_after_if[col]
    seq1 = (seq1[seq1 != MISSING]) % PERIOD_ZERO
    seq2 = (seq2[seq2 != MISSING]) % PERIOD_ZERO
    if len(seq1) == 0 or len(seq2) == 0:
        continue
#    print(seq1)
#    print(seq2)
    before_and_after_deltas.append(int(round(seq2.mean() - seq1.mean()) % PERIOD_ZERO))
all_deltas = list(set(before_and_after_deltas))
if len(all_deltas) != 1:
    print("warning: there is not a unique delta for if-branch. use majority vote instead?")
PERIOD_SHIFT = all_deltas[0]
# Prescience mode on. We peaked at the data manually.
period_shift = 33
if period_shift != PERIOD_SHIFT:
    print("warning: unexpected period shift")
print("Detected period shift for if-branch:", PERIOD_SHIFT)
```

```python
# Now let's model the location of the PoIs for sanity checks.
# Note that this can not be a simple linear regression, because there is a second order effect
# due to bit packing. 8 columns are packed in one byte, and accessing columns for the
# check if the matrix is systematic takes time depending on col%8.
# This is why we only take every 8th column here, and correct for that later.
COL_TIMESTAMPS_STEP=8
col_timestamps = []
fit_col_timestamps = []
for col in range(0, params.n):
    seq = seqs_before_if[col]
    if len(seq) == 0 or seq[0] == MISSING:
        continue
    col_timestamps.append([col, seq[0]])
    if col % COL_TIMESTAMPS_STEP == 0:
        fit_col_timestamps.append([col, seq[0]])

col_timestamps = np.array(col_timestamps)
fit_col_timestamps = np.array(fit_col_timestamps)

# Intercept, slope, and acceleration
var('ci', 'cs', 'ca')
model(x) = ca*x**2 + cs*x + ci
fit_result = find_fit(fit_col_timestamps, model)
fit_values = {}
for eq in fit_result:
    fit_values[eq.left()] = round(eq.rhs(), 4)
# Result is: {ca: -315.0, ci: 1047823.0, cs: 2264764.75}
first_approx = model(ca=fit_values[ca], cs=fit_values[cs], ci=fit_values[ci])

# Now do a linear fit on col % COL_TIMESTAMP_STEP
# Can't use 0-8 because our full before-seqs start at 5. So we shift from 8 down to 0.
fit_diff = np.array([[col-COL_TIMESTAMPS_STEP, ts-first_approx(x=col)] for col,ts in col_timestamps])
fit_diff=fit_diff[(fit_diff[:,0] >= 0) & (fit_diff[:,0] < COL_TIMESTAMPS_STEP)]
fit_diff = fit_diff.astype(np.float64)
var('mci', 'mcs')
model_diff(x) = mcs*x + mci
fit_result_diff = find_fit(fit_diff, model_diff)
fit_values_diff = {}
for eq in fit_result_diff:
    fit_values_diff[eq.left()] = round(eq.rhs(), 4)
# Result is: {ca: -315.0, ci: 1047823.0, cs: 2264764.75}
second_approx = model_diff(mci=fit_values_diff[mci], mcs=fit_values_diff[mcs])

def model_start_ts_col(col):
    return int(first_approx(x=col) + second_approx(x=col%8))

diff_to_model = [model_start_ts_col(c) for c in range(params.n - col_timestamps.shape[0], params.n)] - col_timestamps[:,1]
if (diff_to_model != 0).any():
    print("Model of start of zero stage columns differs for some values")

# Find the PoI for zero stage processing column col and row.
def model_poi_ts(col, row):
    if row == col:
        return MISSING
    elif row < col:
        return model_start_ts_col(col) + row * PERIOD_ZERO
    else: # row > col
        return model_start_ts_col(col) + (row - 1) * PERIOD_ZERO + PERIOD_SHIFT
```

```python
# Let's solve the binary classification problem for the leakpos bits that are not missing.
# Collect all sample values for 0 and 1 for each leakpos.
non_missing_positions = np.where(leakpos != MISSING)[0]
considered_leakpos = leakpos[non_missing_positions]
leak_data = np.zeros((1000, len(considered_leakpos)), dtype=np.uint8)
trace_data = np.zeros((1000, len(considered_leakpos)))
for t in tqdm(range(tr.count)):
    fname_base = tr.trace_fname_base(t)
    key = McElieceKey(fname_base + '.nfo')
    leak_data[t] = np.frombuffer(key.leak[:(PK_NROWS-1)*params.n], dtype=np.uint8)[non_missing_positions]

    trace = np.load(fname_base + '.npy')
    trace_data[t] = trace[considered_leakpos]
    del trace
```

```python
leak0_mean = np.zeros(len(considered_leakpos))
leak0_var = np.zeros(len(considered_leakpos))
leak0_size = np.zeros(len(considered_leakpos))
leak1_mean = np.zeros(len(considered_leakpos))
leak1_var = np.zeros(len(considered_leakpos))
leak1_size = np.zeros(len(considered_leakpos))
mask0 = (leak_data == 0)
mask1 = (leak_data == 1)
for idx, pos in tqdm(list(enumerate(considered_leakpos))):
    leak0_size[idx] = np.sum(leak_data[:,idx])
    leak1_size[idx] = 1000-leak0_size[idx]
    col0 = trace_data[:, idx][mask0[:, idx]]
    col1 = trace_data[:, idx][mask1[:, idx]]
    leak0_mean[idx] = np.mean(col0)
    leak0_var[idx] = np.std(col0)
    leak1_mean[idx] = np.mean(col1)
    leak1_var[idx] = np.std(col1)
    
```
